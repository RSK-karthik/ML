```python
import pandas as pd
import numpy as np
df=pd.read_csv('employees.csv')
df = pd.read_csv("employees.csv", usecols=['EMPLOYEE_ID','MANAGER_ID','DEPARTMENT_ID'])
```

```python
from sklearn.impute import SimpleImputer
imputer =SimpleImputer(missing_values=np.nan,strategy='mean')
imputer = imputer.fit(df[['DEPARTMENT_ID']])
df[['DEPARTMENT_ID']] = imputer.transform(df[['DEPARTMENT_ID']])
df
```

```python
from sklearn.impute import SimpleImputer
imputer =SimpleImputer(missing_values=np.nan,strategy='median')
imputer = imputer.fit(df[['MANAGER_ID']])
df[['MANAGER_ID']] = imputer.transform(df[['MANAGER_ID']])
df
```

```python
from sklearn.impute import SimpleImputer
imputer =SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer = imputer.fit(df[['DEPARTMENT_ID']])
df[['DEPARTMENT_ID']] = imputer.transform(df[['DEPARTMENT_ID']])
df
```

```python
#2
hotEncoding=pd.get_dummies(df,columns=["MANAGER_ID"])
hotEncoding.head()
```

```python
#3
#L1 Norm
df=pd.read_csv('index.csv')
maximum=df["Weight"].max()
df["L1_Norm_Weight"]=df["Weight"]/maximum
```

```python
#L2 Norm

summation=df["Weight"].sum()
df["L2_Norm_Weight"]=df["Weight"]/summation
```

```python
#Max

squareSum=(df["Weight"]**2).sum()
df["Max_Norm_Weight"]=df["Weight"]/(np.sqrt(squareSum))
```
